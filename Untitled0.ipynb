```python
# Advanced Technical Quiz System (Python)

import random
import time

questions = {
    "easy": [
        ("What does CPU stand for?",
         ["Central Processing Unit", "Computer Power Unit", "Central Program Unit", "Control Processing Unit"], "A"),
        ("Which language is used for web pages?",
         ["Python", "C++", "HTML", "Java"], "C")
    ],
    "medium": [
        ("What is Machine Learning?",
         ["Programming manually", "Machines learning from data", "Hardware technology", "None"], "B"),
        ("Which data structure uses FIFO?",
         ["Stack", "Queue", "Tree", "Graph"], "B")
    ],
    "hard": [
        ("Which algorithm is used for classification?",
         ["K-Means", "Naive Bayes", "Apriori", "PCA"], "B"),
        ("What is overfitting?",
         ["Good model", "Model fits noise", "Less data", "High accuracy"], "B")
    ]
}

print("=== ADVANCED TECHNICAL QUIZ SYSTEM ===\n")

level = input("Choose difficulty (easy / medium / hard): ").lower()
quiz_questions = random.sample(questions[level], len(questions[level]))

score = 0
start_time = time.time()

for i, q in enumerate(quiz_questions, 1):
    print(f"\nQ{i}. {q[0]}")
    options = q[1]
    for idx, opt in enumerate(options):
        print(f"{chr(65+idx)}. {opt}")

    question_start = time.time()
    answer = input("Your answer (A/B/C/D): ").upper()
    question_time = time.time() - question_start

    if question_time > 15:
        print("Time Over ‚è∞")
    elif answer == q[2]:
        print("Correct ‚úÖ")
        score += 1
    else:
        print(f"Wrong ‚ùå Correct answer: {q[2]}")

end_time = time.time()

accuracy = (score / len(quiz_questions)) * 100

print("\n=== QUIZ RESULT ===")
print(f"Score        : {score}/{len(quiz_questions)}")
print(f"Accuracy     : {accuracy:.2f}%")
print(f"Total Time   : {end_time - start_time:.2f} seconds")

if accuracy >= 80:
    print("Grade: Excellent üéâ")
elif accuracy >= 50:
    print("Grade: Good üëç")
else:
    print("Grade: Needs Improvement üìò")
```
